```python
#spiral array
def numpy_spiral_order(matrix):
    spiral = []
    top, left = 0, 0
    bottom, right = matrix.shape[0] - 1, matrix.shape[1] - 1

    while top <= bottom and left <= right:
        for i in range(left, right + 1):
            spiral.append(matrix[top, i])
        top += 1
        
        for i in range(top, bottom + 1):
            spiral.append(matrix[i, right])
        right -= 1

        if top <= bottom:
            for i in range(right, left - 1, -1):
                spiral.append(matrix[bottom, i])
            bottom -= 1

        if left <= right:
            for i in range(bottom, top - 1, -1):
                spiral.append(matrix[i, left])
            left += 1

    return spiral

spiral_order = numpy_spiral_order(arr)
print("Spiral Order Traversal:", spiral_order)
```

```python
#videogames 
import pandas as pd 
import matplotlib.pyplot as plt 

data = pd.read_csv('videogame.csv')
#1
data['global_sales'] = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum(axis=1)
data = data.sort_values(by='global_sales', ascending=False)
print("Sorted DataFrame by Global Sales:\n", data.head())

#Plot1
genre_sales = data.groupby('Genre')['global_sales'].sum()
genre_sales.plot(kind='bar', title='Global Sales by Genre', xlabel='Genre', ylabel='Total Sales', color='red')
plt.show()

#GTA
gta_games = data[data['Name'].str.contains('Grand Theft Auto', na=False)]
gta_summary = gta_games[['Name', 'Platform', 'Year', 'EU_Sales', 'JP_Sales']].copy() 
gta_summary['EU_JP_Sales_Sum'] = gta_summary[['EU_Sales', 'JP_Sales']].sum(axis=1)  
print("Grand Theft Auto Games Summary")
print(gta_summary)


# Pie chart for GTA sales
sales_regions = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
gta_sales_totals = gta_games[sales_regions].sum()
gta_sales_totals.plot(kind='pie', title='GTA Total Sales by Region')
plt.ylabel('')
plt.show()
```

```python
#Part1 

import numpy as np
import matplotlib.pyplot as plt

#Beta
a, b = 4, 20
beta_samples = np.random.beta(a, b, 10000) * 100

#Exponential
scale_exp = 0.1
exponential_samples = np.random.exponential(scale_exp, 10000) * 100

#Gamma
shape_gamma, scale_gamma = 2, 0.1
gamma_samples = np.random.gamma(shape_gamma, scale_gamma, 10000) * 100

#Laplace
loc_laplace, scale_laplace = 0, 0.5
laplace_samples = np.random.laplace(loc_laplace, scale_laplace, 10000) * 100

#Normal
loc_normal, scale_normal = 0, 3
normal_samples = np.random.normal(loc_normal, scale_normal, 10000)

#Poisson
lambda_poisson = 3
poisson_samples = np.random.poisson(lambda_poisson, 10000)

# Plotting all distributions together
fig, axes = plt.subplots(3, 2, figsize=(12, 18))

# Beta Distribution
axes[0, 0].hist(beta_samples, bins=np.arange(-5, 51, 1), alpha=0.75, color='blue')
axes[0, 0].set_title("Beta Distribution")
axes[0, 0].set_xlabel("Value")
axes[0, 0].set_ylabel("Frequency")

# Exponential Distribution
axes[0, 1].hist(exponential_samples, bins=np.arange(-1, 51, 1), alpha=0.75, color='green')
axes[0, 1].set_title("Exponential Distribution")
axes[0, 1].set_xlabel("Value")
axes[0, 1].set_ylabel("Frequency")

# Gamma Distribution
axes[1, 0].hist(gamma_samples, bins=np.arange(-1, 51, 1), alpha=0.75, color='red')
axes[1, 0].set_title("Gamma Distribution")
axes[1, 0].set_xlabel("Value")
axes[1, 0].set_ylabel("Frequency")

# Laplace Distribution
axes[1, 1].hist(laplace_samples, bins=np.arange(-1, 51, 1), alpha=0.75, color='purple')
axes[1, 1].set_title("Laplace Distribution")
axes[1, 1].set_xlabel("Value")
axes[1, 1].set_ylabel("Frequency")

# Normal Distribution
axes[2, 0].hist(normal_samples, bins=np.arange(-10, 12, 1), alpha=0.75, color='orange')
axes[2, 0].set_title("Normal Distribution")
axes[2, 0].set_xlabel("Value")
axes[2, 0].set_ylabel("Frequency")

# Poisson Distribution
axes[2, 1].hist(poisson_samples, bins=np.arange(-1, 12, 1), alpha=0.75, color='cyan')
axes[2, 1].set_title("Poisson Distribution")
axes[2, 1].set_xlabel("Value")
axes[2, 1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#Part2 

import numpy as np

matrix = np.loadtxt('matrix.csv', delimiter=',') 


#1
# Task 1:
upper_diagonal = np.triu(matrix)
print("Task 1: Upper diagonal matrix (column-wise):")
print(upper_diagonal.flatten())

#Task 2:
mean = np.mean(matrix, axis=0)
median = np.median(matrix, axis=0)
std_dev = np.std(matrix, axis=0)

det = np.linalg.det(matrix)

if det != 0:
    inverse = np.linalg.inv(matrix)
else:
    inverse = np.linalg.pinv(matrix)

print("\nTask 2: Mean, Median, Standard Deviation, Determinant, and Inverse:")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
print(f"Determinant: {round(det, 2)}")
print(f"Inverse: \n{np.round(inverse, 2)}" if det != 0 else f"Pseudo-Inverse: \n{np.round(inverse, 2)}")

#Task 3:
sorted_vertical = np.sort(matrix, axis=0)
sorted_horizontal = np.sort(matrix, axis=1)
flatten_sorted = np.sort(matrix.flatten())

print("\nTask 3: Sorted Matrices:")
print("Sorted vertically:\n", sorted_vertical)
print("Sorted horizontally:\n", sorted_horizontal)
print("Flattened and sorted:\n", flatten_sorted)

#Task 4:
flattened = matrix.flatten()
unique, counts = np.unique(flattened, return_counts=True)
unique_counts = dict(zip(unique, counts))

print("\nTask 4: Unique Elements and Frequencies:")
for key in unique_counts:
    print(f"{key}: {unique_counts[key]}")

second_largest = sorted(unique)[-2]
second_largest_freq = unique_counts[second_largest]
print(f"Frequency of the second largest element: {second_largest_freq}")

# Task 5:
def pad_matrix(matrix, n):
    padded = np.pad(matrix, ((n, n), (n, n)), mode='constant', constant_values=0)
    return padded

# Eg
padded_matrix = pad_matrix(matrix, 2)

print("\nTask 5: Padded Matrix:")
print(padded_matrix)
```
